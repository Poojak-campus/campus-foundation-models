```python
import pandas as pd

train = pd.read_csv('/content/p1_synthetic_train.csv')

print(train.head())
print(train.columns)
print(train.shape)
```

```python
print(train.columns)
```

```python
print(train['mastery_score'].isna().sum())
```

```python
print(train['mastery_score'].head(20))
```

```python
train = train.dropna(subset=['mastery_score'])
```

```python
print(train['mastery_score'].isna().sum())
```

```python
features = [
    'decay_rate',
    'difficulty',
    'response_correct',
    'bloom_tier'
]

print(train[features].isna().sum())
```

```python
from sklearn.ensemble import RandomForestClassifier

X_train = train[features]
y_train = train['mastery_score']

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

print("Training completed")
```

```python
import pandas as pd

test = pd.read_csv('/content/p1_synthetic_test.csv')

print(test.shape)
```

```python
test = test.dropna(
    subset=[
        'decay_rate',
        'difficulty',
        'response_correct',
        'bloom_tier',
        'mastery_score'
    ]
)
```

```python

features = [
    'decay_rate',
    'difficulty',
    'response_correct',
    'bloom_tier'
]
X_test = test[features]
y_test = test['mastery_score']

pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, pred)

print("Accuracy:", acc)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, pred)

print("F1 Score:", f1)
```

```python
importance = pd.DataFrame({
    'Feature': features,
    'Importance': model.feature_importances_
})

importance = importance.sort_values(
    by='Importance',
    ascending=False
)

print("\n===== Feature Importance =====")
print(importance)
```

```python
from sklearn.metrics import classification_report

print("\n===== Classification Report =====")
print(classification_report(y_test, pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)

print("\n===== Confusion Matrix =====")
print(cm)
```
